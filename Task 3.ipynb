```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data= pd.read_excel('tips.xlsx')
data1=pd.read_csv('householdtask3.csv')
```

```python
plt.scatter(data['total_bill'],data['tip'],c='brown',marker='*')
plt.title('Total bill vs Tip ')
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.grid(False)
plt.tight_layout()
```

```python
time_count=data['time'].value_counts()
plt.pie(time_count.values,labels=time_count.index,autopct='%1.1f%%',colors=['red','yellow'])
plt.title('Distibution of Dining time')
plt.axis('equal')
plt.tight_layout()
```

```python
plt.bar(data['smoker'].value_counts().index,data['smoker'].value_counts().values,edgecolor='black')
plt.title('Count of Smokers and Non Smokers')
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.grid(axis='y')
plt.tight_layout()
```

```python
plt.hist(data['tip'],bins=15,edgecolor='black')
plt.title('Distribution of Tip amount')
plt.xlabel('Tip')
plt.ylabel('Frequency')
plt.tight_layout()
```

```python
plt.pie(data['sex'].value_counts().values,labels=data['sex'].value_counts().index,autopct='%1.1f%%',colors=['blue','yellow'])
plt.title('Distibution of Gender')
plt.axis('equal')
plt.tight_layout()
```

```python
plt.hist(data['size'],bins=15,edgecolor='black',color='lightblue')
plt.title('Distribution of Number of people')
plt.xlabel('Tip')
plt.ylabel('Frequency')
plt.tight_layout()
```

```python
plt.bar(data['day'].value_counts().index,data['day'].value_counts().values,edgecolor='black')
plt.title('Count of days that most people go out')
plt.xlabel('Day')
plt.ylabel('Count')
plt.tight_layout()
```

```python
plt.plot(data1['year'], data1['size'], marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Household Size')
plt.title('Household Size Over Years')
plt.grid(True)
plt.tight_layout()
```

```python
plt.plot(data1['year'], data1['eqv_income'], marker='o', linestyle='-', color='m', label='Equivalized Income')
plt.plot(data1['year'], data1['eqv_exp'], marker='o', linestyle='-', color='c', label='Equivalized Expenditure')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.title('Equivalized Income and Expenditure Over Years')
plt.legend()
plt.grid(True)
plt.tight_layout()
```
